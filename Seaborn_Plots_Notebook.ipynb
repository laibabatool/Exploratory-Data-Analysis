```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

**Explanation:**  
This scatter plot shows the relationship between `total_bill` and `tip`. Points are distinguished by `time` (Lunch or Dinner) using different colors and styles.

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title("Bar Plot of Average Total Bill by Day and Gender")
plt.show()
```

**Explanation:**  
This bar plot displays the average `total_bill` for each day, separated by gender (`sex`). Error bars represent the standard deviation.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title("Box Plot of Total Bill by Day and Gender")
plt.show()
```

**Explanation:**  
The box plot shows the distribution of `total_bill` for each day, with data split by gender. It highlights the median, quartiles, and potential outliers.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.title("Violin Plot of Total Bill by Day and Gender")
plt.show()
```

**Explanation:**  
The violin plot combines a box plot and a kernel density estimate, showing the distribution of `total_bill` for each day, separated by gender.

```python
# Pair plot with KDE diagonal
sns.pairplot(data=tips, hue='sex', diag_kind='kde')

# Add title
plt.suptitle("Pair Plot of Tips Dataset", y=1.02)

# Display the plot
plt.show()
```

**Explanation:**  
This pair plot visualizes pairwise relationships between numerical columns (`total_bill`, `tip`, etc.), with histograms on the diagonal and points colored by gender.

```python
plt.figure(figsize=(8, 6))
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**Explanation:**  
This heatmap shows the correlation matrix of numerical variables, with color intensity indicating the strength and direction of correlations.

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=tips, x=tips.index, y='total_bill', hue='time')
plt.title("Line Plot of Total Bill Over Time")
plt.show()
```

**Explanation:**  
This line plot displays the trend of `total_bill` over time, with separate lines for Lunch and Dinner.

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=tips, x='day', hue='sex')
plt.title("Count Plot of Gender Distribution by Day")
plt.show()
```

**Explanation:**  
The count plot shows the number of occurrences of `sex` (gender) for each day of the week.

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=tips, x='total_bill', hue='time', fill=True, alpha=0.5)
plt.title("KDE Plot of Total Bill by Time")
plt.show()
```

**Explanation:**  
This kernel density estimate (KDE) plot visualizes the distribution of `total_bill` for Lunch and Dinner, with shaded areas representing density.
